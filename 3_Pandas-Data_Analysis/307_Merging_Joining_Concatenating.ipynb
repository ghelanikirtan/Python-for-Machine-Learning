# 1. Data Frames: (For Concatenation)

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

# Concatenation:
- **It glues together DataFrames.**
- **Dimensions should be same along the axis to be concatenated.**
- **`pd.concat` can be used to pass list of DataFrames to concatenate together.**

```python
df_list = [df1, df2, df3]
pd.concat(df_list)
#Note: here Columns are equal cause dataFrames are conctenated along the rows
#      i.e. axis=0
```

```python
pd.concat(df_list, axis=1)
```

```python
# using map and lambda functions on dataframes

pd.concat(list(map(lambda n : n.T, df_list)), axis=1)
```

# 2. DataFrames : (For Merging) 

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left
```

```python
right
```

# Merging:
- **Syntax: `pd.merge(left_df,right_df, how='X', on='keyValue'/listOfkeys)`**
- **where, `X = inner/outer/left/right/cross` - though not much useful**

```python
pd.merge(left, right, how='inner', on='key')
```

```python
# other EXAMPLE :
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key1')
# Output is the reason why 'on' is important
```

```python
pd.merge(left, right, on=['key1','key2'])

# default how='inner', Use intersection of keys from both frames
```

> **Extras :**

```python
pd.merge(left, right, how='outer', on=['key1','key2'])

# Outer - Use union of keys from both frames
```

```python
pd.merge(left, right, how='right', on=['key1','key2'])

# right - Use keys from right frame only
```

```python
pd.merge(left, right, how='left', on=['key1','key2'])

# left - Use keys from left frame only
```

# 3. DataFrames : (For Joining) 

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

# Joining:
- **Joining is convenient method for combining the columns of two potentially differently-indexed Data Frames into single result DataFrame.**
- **Similar to 'merge', except they actually are joining on the index instead of column.**
- **Syntax: `df_name_1.join(df_name_2, how='xD')`**

```python
left.join(right)
```

```python
right.join(left)
```

```python
left.join(right, how='outer')
```
