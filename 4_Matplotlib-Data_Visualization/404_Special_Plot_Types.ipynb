# Special Plot Types:
**There are many specialized plots as follows:**
- **barplots**
- **histograms**
- **scatter plots**
- **boxplot**
- **and many more...**

Mostly, seaborn (a statistical plotting library for python) is used for such kind of plots.

```python
import matplotlib.pyplot as plt
import numpy as np
from random import sample
import math
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2

print("x : ",x,"\ny : ",y)
```

___

## 1. Scatter Plots:

```python
plt.scatter(x,y)
```

```python
data = sample(range(1,1000), 100)
plt.scatter(data, sample(range(1,1000), 100))
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.scatter(data, [np.random.normal(0, std, 100) for std in range(1,4)])
```

___

## 2. Histograms:

```python
data = sample(range(1,1000), 100)
plt.hist(data)
```

___

## 3. boxplot:

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

fig, axes = plt.subplots(1,2, figsize=(12,4))

# Rectangular box plot
axes[0].boxplot(data, vert=True, patch_artist=True)
axes[1].boxplot(data)
```
