# Functional Programming with MatplotLib:

### Basic Use case:

>**`figureName, axesName = plt.subplots(nrows = rowNumber, ncols = colNumber)`**


```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes both
fig,axes = plt.subplots() 
```

```python
fig,axes = plt.subplots() 

# Plotting in created figure
axes.plot(x,y) 
```

### Rows and columns can also be specified while creating the subplots() object:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# more examples:
fig_1, axes_1 = plt.subplots(nrows=3, ncols=3)

# To Fix ovelapping of the two consecutives rows and columns use plt.tight_layout()
plt.tight_layout()
```

**Here, axes and axes_1 are the arrays of axes object (matplotLib axes) So They are iterable like a normal arrays as shown below:**

```python
# 1 x 2 (normal 1-D array)
axes
```

```python
# 3 x 3 array
axes_1
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for curr_axes in axes:
    curr_axes.plot(x,y)
```

```python
fig, axes_1 = plt.subplots(nrows=3,ncols=3)

for i in axes_1:
    for j in i:
        j.plot(x,y)
        
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI:

**Manual Method:**
- **`figureName = plt.figure( figsize=(width, height) , dpi= dpiValue)`**
- **`axes = fig.add_axes([0, 0, 1, 1])`**
- **`axes.plot(x,y)`**

**Tuple Unpacking Method:**
- **`figureName, axesName = plt.subplots( nrows=rowNumber, ncols=colNumber, figsize=(width, height) , dpi= dpiValue )`**
- **`axesName/[indexing].plot(x,y)`**

**NOTE:**
- Here, dpi means dots per inch

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4), dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Legend:


- In `axes.plot(**args, label='labelName')` and use **`axes.legend(loc = LocationCode{default:0})`** at the end.
- The legend function takes an optional keyword argument loc that can be used to specify where in the figure the legend is to be drawn.See the **[Documentation Page](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html?highlight=axes%20legend#matplotlib.axes.Axes.legend)** for more details.

```python
fig, axes = plt.subplots(figsize=(5,5))
axes.plot(x,x**2)
axes.plot(x,x**3)

axes.legend()
```

```python
fig, axes = plt.subplots(figsize=(5,5))
axes.plot(x,x**2, label = 'x squared')
axes.plot(x,x**3, label = 'x cubed')

axes.legend()
```

```python
fig.savefig('Image_data/My_Picture_402001.png')
```

```python
# Lots of options....

axes.legend(loc=1) # upper right corner
axes.legend(loc=2) # upper left corner
axes.legend(loc=3) # lower left corner
axes.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
axes.legend(loc=0) # let matplotlib decide the optimal location
fig
```
