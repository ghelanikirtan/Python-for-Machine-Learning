# MatplotLib Commands :

**Two ways of creating matplotlib plots:**
1. **Functional Method.**
2. **Object Oriented Method.**

___

# 1. Functional Method:

**`plt.plot(x_coordinate, y_coordinate ...n, 'color & line styles attributes')`**

```python
plt.plot(x,y)
# plt.show() - on another editor this is mandatory 
```

```python
# Labelling axis in using .xlabel(' ') & .ylabel(' ')
plt.plot(x,y)
plt.xlabel('X - Label Name here')
plt.ylabel('Y - Label Name here')

# Giving Title to plot using .title(' ')
plt.title('Title Name here')
```

## Creating Multiple plots on the same canvas : 

**`plt.subplot( no_of_Rows, no_of_Cols, Plot_Number )`**

```python
# creates empty cell 
plt.subplot(1,2,1)
plt.plot(x,y, 'r-.')         # Plotting on that cell

plt.subplot(1,2,2)
plt.plot(y,x, 'b*-')
```

# 2. Object Oriented Method:

- In this we will instantiate figure objects and then call methods or attributes from that object.

- The main idea in using the more formal Object Oriented Method is to **create figure objects and then just call methods or attributes of that object.**

- This approach is better while dealing with a canvas having multiple plots on it.

**`figureName = plt.figure()`**<br>
**`axesName = figureName.add_axes([left, bottom, width, height])`**<br>
**`axesName.plot(*args, attributes)`**<br>

```python
# Creating a Figure (EMPTY CANVAS)
fig = plt.figure()

# adding set of axes to figure 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
# passing a list of 4 arguments - left, bottom, width, height (range 0 to 1)

# plot on that set of axes
axes.plot(x,y, 'r')

# set labels and title in Object Oriented method
axes.set_xlabel('X - Label name here')
axes.set_ylabel('Y - Label name here')
axes.set_title('Title Name here')
```

___

## Creating Multiple Figures in same canvas using OO:

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #here, args in lists are in relation with axes1

# Plotting
axes1.plot(x,y)
axes1.set_title('PARENT PLOT')

axes2.plot(y,x, 'r')
axes2.set_title('CHILD PLOT')
axes2.set_xlabel('X - Label Axes2')
axes2.set_ylabel('Y - Label Axes2')
```
