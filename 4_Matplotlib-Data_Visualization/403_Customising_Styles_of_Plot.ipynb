# Customising the Styles of Plot:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2

print("x : ",x,"\ny : ",y)
```

---

# Plot Appearance:

### 1. Setting Colors:
- **`axes.plot(x,y, color='colorString')`** - colorString can be a hexcode or color name 


```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# axes.plot(x,y, color='r')
# axes.plot(x,y, color='purple')
axes.plot(x,y, color='#FF8C00') #RGB Hex code
```

### 2. Line width and Transparency:
- **`axes.plot(x,y, color='colorString', linewidth=1, alpha=1)`** 
- **'linewidth' OR 'lw' takes integer/decimal value and sets *width* of line accordingly.**
- **'alpha' takes integer/decimal value (range[0,1]) and sets *transparency* of the line accordingly.**

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=3.5, alpha=0.5)
```

### 3. Line Styles and Markers:
- **`axes.plot(x,y, linestyle='lineStyleCharacters', marker='char', markersize='integerValue')`** 
- **'linestyle' OR 'ls' takes character value as input and decides line style accordingly.**
- **'marker' takes character value and mark the points in figure to be plotted.**
- **For more reference on marker [click here](https://matplotlib.org/3.5.1/gallery/lines_bars_and_markers/marker_reference.html).**

```python
fig = plt.figure(figsize=(12,6))
axes = fig.add_axes([0,0,1,1])

# Line Styles:
axes.plot(x,x-1 , color='brown', lw=2, linestyle='--')
axes.plot(x,x, color='blue', lw=2, linestyle='-.')
axes.plot(x,x+1 , color='red', lw=2, linestyle=':')
axes.plot(x,x+2 , color='black', lw=3, linestyle='-')

# marker amd markersize:
axes.plot(x,x+3 , color='red', lw=0, ls='-', marker='o') #common ones = 'o', '+', '*', '.', '1', '2', '3', '4', ...
axes.plot(x,x+4 , color='blue', lw=1, ls='-', marker='+', )
axes.plot(x,x+5 , color='green', lw=1, ls='-', marker='*')
axes.plot(x,x+6 , color='violet', lw=0, ls='-', marker='1', markersize=10)
axes.plot(x,x+7 , color='purple', lw=0, ls='-', marker='2', markersize=15)

# markerfacecolor, markeredgewidth, and markeredgecolor:
axes.plot(x,x+7 , color='purple', lw=2, ls='-', marker='o', markersize=20, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='blue')
```

```python
# More on linewidth, linestyles and markers

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

# Control over axis appearance:

### Plot Range:
- The ranges of the axes can be configure using **`.set_ylim([from, to])`** and **`.set_ylim([from, to])`** methods in the axis object.
- **`.axis('tight')`** method in the axis object can be used for automatically getting **"tightly fitted"** axes ranges. 

```python
fig, axes = plt.subplots()

axes.plot(x,y, color='purple', lw=2, ls='--')

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```

```python
fig, axes = plt.subplots()

axes.plot(x,y, color='purple', lw=2, ls='--')

axes.axis('tight') # automatically getting "tightly fitted" axes ranges
```

```python
# More examples: (Comparison)

fig, axes = plt.subplots(1,3, figsize=(12,4))

# plotting more than one in single function/method (likewise inline)
# axes[0].plot(x, x**2, x, x**3, x, x**4)
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range")
```
